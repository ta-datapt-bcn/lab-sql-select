## Challenge 1

```python
query1 = """
SELECT
    authors.au_id,
    authors.au_fname,
    authors.au_lname,
    titles.title,
    publishers.pub_name

FROM
    `ironhack-data-analytics.publications.authors` authors
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor` title_author
USING (au_id)

INNER JOIN
    `ironhack-data-analytics.publications.titles` titles
ON
    title_author.title_id = titles.title_id

INNER JOIN
    `ironhack-data-analytics.publications.publishers` publishers
USING (pub_id)

ORDER BY
    au_id

"""
```

```python
query_job = client.query(query=query1)
df = query_job.to_dataframe()
df.head(20)
```

## Challenge 2

```python
query1 = """
SELECT
    authors.au_id,
    authors.au_fname,
    authors.au_lname,
    publishers.pub_name,
    COUNT(titles.title) AS title_count

FROM
    `ironhack-data-analytics.publications.authors` authors
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor` title_author
ON
    authors.au_id = title_author.au_id

INNER JOIN
    `ironhack-data-analytics.publications.titles` titles
ON
    title_author.title_id = titles.title_id

INNER JOIN
    `ironhack-data-analytics.publications.publishers` publishers
ON
    titles.pub_id = publishers.pub_id

GROUP BY
    1,2,3,4

ORDER BY
    title_count DESC,
    pub_name

"""
```

```python
query_job = client.query(query=query1)
df = query_job.to_dataframe()
df
```

## Challenge 3

```python
query3 = """
SELECT
    a.au_id,
    a.au_lname,
    a.au_fname,
    SUM(s.qty) AS total

FROM
    `ironhack-data-analytics.publications.authors` a
LEFT JOIN
    `ironhack-data-analytics.publications.titleauthor` t_a
USING(au_id)

LEFT JOIN
    `ironhack-data-analytics.publications.sales` s
USING(title_id)

GROUP BY
    a.au_id,
    a.au_lname,
    a.au_fname

ORDER BY
    total DESC

LIMIT
    3

"""
```

```python
query_job = client.query(query=query3)
df = query_job.to_dataframe()
df.head(20)
```

## Challenge 4

```python
query4 = """
SELECT
    a.au_id,
    a.au_lname,
    a.au_fname,
    COALESCE((SUM(s.qty)),0) as total

FROM
    `ironhack-data-analytics.publications.authors` a
LEFT JOIN
    `ironhack-data-analytics.publications.titleauthor` t_a
USING(au_id)

LEFT JOIN
    `ironhack-data-analytics.publications.sales` s
USING(title_id)

GROUP BY
    a.au_id,
    a.au_lname,
    a.au_fname

ORDER BY
    total DESC

"""
```

```python
query_job = client.query(query=query4)
df = query_job.to_dataframe()
df
```
