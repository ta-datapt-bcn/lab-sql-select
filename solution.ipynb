```python
%%sql

SELECT title_id, sum(qty) as qty
FROM sales
GROUP BY title_id
```

```python
%%sql

--Hay autores que no tienen libros publicados

SELECT  authors.au_id as 'AUTHOR_ID', authors.au_lname as 'LAST_NAME', 
        authors.au_fname as'FIRST_NAME', titles.title as 'TITLE', 

    case 
       when sum(sales.qty) is NULL then 0
       else sum(sales.qty) 
    end as 'TOTAL'
      
FROM authors LEFT JOIN titleauthor ON authors.au_id = titleauthor.au_id
LEFT JOIN titles ON titles.title_id = titleauthor.title_id

LEFT JOIN   (SELECT title_id, sum(qty) as qty
            FROM sales
            GROUP BY title_id) as sales 

ON titles.title_id = sales.title_id

GROUP BY AUTHOR_ID
ORDER BY TOTAL DESC
```

```python
%%sql
select authors.au_id as 'AUTHOR ID', authors.au_lname as 'LAST NAME', 
        authors.au_fname as 'FIRST NAME', titles.title as 'TITLE',
        publishers.pub_name as 'PUBLISHER'
from ((authors inner join titleauthor on authors.au_id = titleauthor.au_id) 
inner join titles on titleauthor.title_id = titles.title_id) 
inner join publishers on titles.pub_id = publishers.pub_id 
LIMIT 10
```

# Challenge 2 - Who Have Published How Many At Where?

```python
%%sql

SELECT authors.au_id as 'AUTHOR_ID', authors.au_lname as 'LAST_NAME', 
authors.au_fname as 'FIRST_NAME', publishers.pub_name as 'PUBLISHER', COUNT(publishers.pub_name) as 'TITLE COUNT'
                   
FROM ((authors inner join titleauthor on authors.au_id = titleauthor.au_id) 
INNER JOIN titles on titleauthor.title_id = titles.title_id) 
INNER JOIN publishers on titles.pub_id = publishers.pub_id 

GROUP BY AUTHOR_ID, PUBLISHER
ORDER BY AUTHOR_ID DESC

LIMIT 10
```

# Bonus Challenge - Most Profiting Authors

```python
%%sql

--Hay autores que no tienen libros publicados

SELECT  authors.au_id as 'AUTHOR_ID', authors.au_lname as 'LAST_NAME', 
        authors.au_fname as'FIRST_NAME', 
        sum(((titles.price*sales.qty*titles.royalty)/100)*(titleauthor.royaltyper/100.0) + titles.advance*(titleauthor.royaltyper/100.0)) as PROFIT
        
FROM authors LEFT JOIN titleauthor ON authors.au_id = titleauthor.au_id
LEFT JOIN titles ON titles.title_id = titleauthor.title_id

LEFT JOIN   (SELECT title_id, sum(qty) as qty
            FROM sales
            GROUP BY title_id) as sales 

ON titles.title_id = sales.title_id

GROUP BY AUTHOR_ID
ORDER BY PROFIT DESC
LIMIT 3
```
