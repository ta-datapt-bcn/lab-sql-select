```python
%%sql
select ta.au_id as author_id, au.au_lname as last_name, au.au_fname as first_name, t.title, pub.pub_name as publisher
from titleauthor ta
inner join authors au on ta.au_id = au.au_id
inner join titles t on ta.title_id = t.title_id
inner join publishers pub on pub.pub_id = t.pub_id
```

## Challenge 2 - Who Have Published How Many At Where?

Podemos aprovechar el código del primer apartado simplemente añadiendo una nueva columna en la que volcaremos el output de un *count* y finalmente obteniendo el conteo mediante *groupby*.

```python
%%sql
select ta.au_id as author_id, au.au_lname as last_name, au.au_fname as first_name, t.title, pub.pub_name as publisher, count(t.title) as title_count
from titleauthor ta
inner join authors au on ta.au_id = au.au_id
inner join titles t on ta.title_id = t.title_id
inner join publishers pub on pub.pub_id = t.pub_id
group by author_id
```

## Challenge 3 - Best Selling Authors

Aprovechando la tabla de autores previamente creada ejecutamos un join con la tabla *sales*, a través del cual podremos obtener el número de ventas por autor. Después procedemos a ordenar la tabla por ventas y mostramos sólo los 3 autores top.

```python
%%sql
select ta.au_id as author_id, au.au_lname as last_name, au.au_fname as first_name, sum(s.qty) as total
from titleauthor ta
inner join authors au on ta.au_id = au.au_id
inner join sales s on s.title_id = ta.title_id
group by author_id
order by total desc
limit 3
```
